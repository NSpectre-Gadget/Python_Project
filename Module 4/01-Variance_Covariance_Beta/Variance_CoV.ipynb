```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
daily_returns_amzn_spx=pd.read_csv(Path("amzn_sp500_data.csv"), index_col='date', parse_dates=True, infer_datetime_format=True)
```

```python
daily_returns_amzn_spx.pct_change().dropna()
```

```python
# Plotting the daily returns for Amazon and the S&P 500
daily_returns_amzn_spx.plot(figsize=(10, 7), title="Daily Returns - Amazon & S&P 500")
```

```python
# Using the Pandas 'var' function, calculate the variance of AMZN and the S&P 500
daily_returns_amzn_spx.var()
```

```python
import pandas as pd

stocks = pd.DataFrame({
'stock_hij' : [142.13, 143.54, 142.52, 143.43, 141.73, 137.68, 139.71, 139.60, 139.61, 141.57],
'stock_klm' : [53.88, 52.83, 52.55, 53.84, 54.16, 55.58, 53.43, 52.50, 53.90, 47.86],
'market' : [3534.22, 3511.93, 3488.67, 3483.34, 3483.81, 3426.92, 3443.12, 3435.56, 3453.49, 3449.52]
})
```
